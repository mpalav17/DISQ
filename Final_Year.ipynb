```python
date = pd.DataFrame()
datedf = pd.read_csv('pre.csv')
date['Date'] = datedf['Date']
date.head(50)
```

```python
list_month = []
for i in range(0,364):
    str = date["Date"][i]
    list_month.append(str[3:5])
print(list_month)
```

```python
finaldf = pd.read_csv('FINAL.csv')
finaldf['Month'] = list_month
finaldf['Month'] = finaldf['Month'].astype(int)
finaldf
```

```python
sdf = finaldf.groupby(['Month']).PrecipitationSumMm.sum().reset_index()
sdf
```

```python
monthlydf = pd.read_csv('Monthly rainfall data.csv').drop([0,1,2])
monthlydf.to_csv('month.csv', index = False)
```

```python
df = pd.read_csv('month.csv')
df.columns = df.values[0]
```

```python
df = df.drop([0])
df
```

```python
df.loc[98] = list_2016
df
```

```python
ndf = df.drop(['SUBDIVISION'], axis=1)
ndf = ndf.astype(float)
```

```python
yearndf = ndf['YEAR']
nndf = ndf.drop(['YEAR'],axis=1)
nndf.values[0]
```

```python
nndf = nndf.fillna(0)
nndf
```

```python
list_annual = []
for i in range(0,98):
    t = 0
    for j in range(0,13):
        t = t + nndf.values[i][j]
    list_annual.append(t)
print(list_annual)
```

```python
nndf['ANNUAL'] = list_annual
nndf
```

```python
nndf .to_csv('final_append.csv', index=False)
```
