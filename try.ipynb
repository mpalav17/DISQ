```python
import pandas as pd
import sklearn as sk
import numpy as np
```

```python
df = pd.read_csv('Rainfall data for the year 2016.csv')
df.head()
```

```python
df = df.replace('T', 0)
df = df.replace('-', 0)
```

```python
df.head()
```

```python
df = df.drop('Date', axis = 1)
df.head()
```

```python
df.shape
```

```python
list_hrs = []
for i in range(0, 364):
    list_hrs.append(i*24)
print(len(list_hrs))
```

```python
df['hrs'] = list_hrs
df.head(336)
```

```python
print(df.values[0])
```

```python
nf = df[335:364]
nf.head()
```

```python
nf.to_csv('needtopredict.csv', index = False)
```

```python
df = df.drop(df.index[335:364])
df.shape
```

```python
list_col = df.columns
for i in range(0, len(list_col)):
    df[list_col[i]] = df[list_col[i]].astype(float)
print(df.values[0])
```

```python
df.to_csv('train_data.csv', index = False)
```

```python
from sklearn.linear_model import LinearRegression as LR
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range = (-1, 1))
```

```python
x = df.drop(['PrecipitationSumInches', 'hrs'], axis = 1)
y = df['PrecipitationSumInches']
```

```python
print(y.values)
```

```python
x = x.values
y = y.values
```

```python
x_train = x[0:300]
y_train = y[0:300]
x_test = x[300:]
y_test = y[300:]
print(len(x_train), len(x_test))
```

```python
x_train = scaler.fit_transform(x_train) 
x_test = scaler.fit_transform(x_test)
```

```python
lr = LR()
```

```python
lr.fit(x_train, y_train)
```

```python
pre_y = lr.predict(x_test)
```

```python
pf = pd.DataFrame({'Actual': y_test, 'Predicted': pre_y})
pf.head(35)
```

```python
print("Accuracy of the model =",lr.score(x_train, y_train))
```
