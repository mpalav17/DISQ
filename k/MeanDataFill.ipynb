```python
from sklearn.model_selection import train_test_split
```

```python
scaler = MinMaxScaler()
x = df.drop('PrecipitationSumInches',axis = 1)
x = scaler.fit_transform(x)
x = pd.DataFrame(x)
x.describe()
```

```python
y = df['PrecipitationSumInches'].multiply(25.4)
y = y[:,None]
y = scaler.fit_transform(y)
y = pd.DataFrame(y)
y.describe()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y,test_size = 0.25, random_state = 42)
x_train.shape
```

```python
x_test.shape
```

```python
y_test = np.array(y_test)
y_test = y_test.reshape(-1,1)
y_test.shape
```

```python
from sklearn.ensemble import RandomForestRegressor# Instantiate model with 1000 decision trees
rf = RandomForestRegressor(n_estimators = 2000, random_state = 42)# Train the model on training data
rf.fit(x_train, y_train)
```

```python
predictions = rf.predict(x_test)# Calculate the absolute errors
pre = scaler.inverse_transform(predictions.reshape(-1,1))
test = scaler.inverse_transform(y_test)
errors = abs(pre - test)# Print out the mean absolute error (mae)
print('Mean Absolute Error:', round(np.mean(errors), 2), 'degrees.')
```

```python
pf = pd.DataFrame()
```

```python
mape = 100 * (errors / test)# Calculate and display accuracy
accuracy = 100 - np.mean(mape)
print('Accuracy:', round(accuracy, 2), '%.')
```
