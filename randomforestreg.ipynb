```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('pre.csv')
df = df.replace('-', 0)
df = df.replace('T', 0)
df = df.drop(df.index[334:365])
df = df.drop('Date', axis=1)
df.shape
#df['PrecipitationSumInches'].values[0]
```

```python
print(np.mean(df['DewPointHighF'].astype(float)))
print(np.median(df['DewPointHighF'].astype(float)))
```

```python
list = df.columns
```

```python
for i in list:
    df[i] = df[i].astype(float)
```

```python
x = df.drop('PrecipitationSumInches',axis=1)
y = df['PrecipitationSumInches']
y = y[:,None]
print(y)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
xt = scaler.fit_transform(x)
yt= scaler.fit_transform(y)
print(yt)
x.shape
yi=scaler.inverse_transform(yt)
print(yi)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split# Instantiate model with 1000 decision trees
x_train, x_test, y_train, y_test = train_test_split(xt, yt, test_size = 0.25)
rf = RandomForestRegressor(n_estimators = 1000, random_state = 42)# Train the model on training data
rf.fit(x_train, y_train)
```

```python
predictions = rf.predict(x_test)# Calculate the absolute errors
errors = abs(predictions - y_test)# Print out the mean absolute error (mae)
print('Mean Absolute Error:', round(np.mean(errors), 2), 'degrees.')
```

```python
mape = 100 * (errors / y_test)# Calculate and display accuracy
accuracy = 100 - np.mean(mape)
print('Accuracy:', round(accuracy, 2), '%.')
```
